# Data Visualization

The `plot()` function in MATLAB helps in producing 2D plots and to get quick help on this function, we can pass the following command :

```matlab
help plot
```

Let's take the following example for plotting :

***Example-1 :***<br>
Plot the location of the ball thrown upwards at every 0.1 seconds until it reaches the ground when the initial velocity is 20 m/s.

````octave
% Initiating variables

u = 20;
g = 9.8;
t = 0;
y = 0;

time = 0;
location = 0;

% Running the loop

while (y>=0)
    t = t + 0.1;
    y = u * t - (0.5 * g * t^2);
    
    time = [time; t];
    location = [location; y];
end

plot(time, location)
````

#### Formatting the Plot

We can follow the following codes to apply various formatting to our plots :

````octave
% To plot a red dashed line :

plot(time, location, "--r")
````

````octave
% To plot a blue line with circular points :

plot(time, location, "-bo")
````

````octave
% To plot a dash-dot red line :

plot(time, location, "-.g")
````

#### Multiple Plots

We can write the following command to produce multiple plots in the same chart :

````octave
% Defining array

x = [0:0.1:5];

% Plotting the curvers

plot(x, sin(x), "-b", x, cos(x), "-r")
````

Another way to plot multiple curves is :

````octave
% Defining variables

x = [0:0.01:5];
y = [sin(x); cos(x); sin(x)-cos(x)];

% Plotting

plot(x,y)
````

Suppose, we need to add some other plot to our existing chart then, we can do it by passing the `hold on` command, as follows :

```matlab
plot(x,y)
hold on
plot(x,sin(x)+ cos(x))
```

If we don't use the `hold on` command then, our plots will get overwritten by the the newly created plot.*(It doesn't work for Jupyter Notebook)*
