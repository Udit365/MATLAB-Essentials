# Data Visualization

The `plot()` function in MATLAB helps in producing 2D plots and to get quick help on this function, we can pass the following command :

```matlab
help plot
```

Let's take the following example for plotting :

***Example-1 :***<br>
Plot the location of the ball thrown upwards at every 0.1 seconds until it reaches the ground when the initial velocity is 20 m/s.

````octave
% Initiating variables

u = 20;
g = 9.8;
t = 0;
y = 0;

time = 0;
location = 0;

% Running the loop

while (y>=0)
    t = t + 0.1;
    y = u * t - (0.5 * g * t^2);
    
    time = [time; t];
    location = [location; y];
end

plot(time, location)
````

#### Formatting the Plot

We can follow the following codes to apply various formatting to our plots :

````octave
% To plot a red dashed line :

plot(time, location, "--r")
````

````octave
% To plot a blue line with circular points :

plot(time, location, "-bo")
````

````octave
% To plot a dash-dot red line :

plot(time, location, "-.g")
````

#### Multiple Plots

We can write the following command to produce multiple plots in the same chart :

````octave
% Defining array

x = [0:0.1:5];

% Plotting the curvers

plot(x, sin(x), "-b", x, cos(x), "-r")
````

Another way to plot multiple curves is :

````octave
% Defining variables

x = [0:0.01:5];
y = [sin(x); cos(x); sin(x)-cos(x)];

% Plotting

plot(x,y)
````

Suppose, we need to add some other plot to our existing chart then, we can do it by passing the `hold on` command, as follows :

```matlab
plot(x,y)
hold on
plot(x,sin(x)+ cos(x))
```

If we don't use the `hold on` command then, our plots will get overwritten by the the newly created plot.*(It doesn't work for Jupyter Notebook)*

#### Labelling the Plot :

A plot is said to be lebelled properly, if it provides all the necessary information at a glance. This includes :

1. Labelling the axes
2. Providing legends
3. Providing chart title
4. Proper scaling of axes

##### Labelling the axes :

`xlabel ("axis_name")` : For labelling X-axis<br>
`ylabel ("axis_name")` : For labelling Y-axis

##### Legend :

`legend("name1","name2")` : Labelling the plot

>***Note:*** The legend names must be in accordance with the order of execution of plots.

##### Title:

`title("chart_title")` : Providing chart title

##### Customized axes
`axis ([0.5 1 -1 1])` : The first two terms defines the lower & upper limits of X-axis and the last two terms are that of for Y-axis.


Let's plot a simple chart before tying out the various labelling options :

````octave
% Initialize the array
t = [0:0.01:0.98];

% Multiple line plot

y1 = sin(2*pi*4*t);
y2 = cos(2*pi*4*t);

y = [y1;y2];

plot(t,y);

%========== Labelling Axes ==============%

% Labelling the X-axis
xlabel ("Time");

% Labelling the Y-axis
ylabel ("Value");

%========== Legend ==============%

legend("sine function","cosine function");

%========== Chart Title ==============%

title("Sine-Cosine Curve");

%========== Customized Axes ==============%

axis ([0.5 1 -1 1]);
````

#### Getting Multiple Figures :

Suppose, we want multiple figures at once (without appending or, overwriting) then, we can do so by using the `figure()` command as follows :

````octave
figure(1);
plot(t,y1,"r");
figure(2);
plot(t,y2,"b");
````

#### Plotting Multiple Figures Side-wise :

Very often we want to compare two charts and the process became easier if we have the plots side-by-side.

This can be achieved using the `subplot()` command that divides the chart area into grids so that, we can plot multiple plots in the same chart side-wise.

The following example shows, how we can split the chart area in 2 parts and do plotting :

````octave
% Dividing plot area into (1x2) -- first 2 arguments & choosing 1st grid to plot (3rd argument)

subplot(1,2,1);
plot(t,y1);

% Choosing in the 2nd grid for plotting

subplot(1,2,2);
plot(t,y2);
````

> ***Note :*** We can perform individual labelling and formatting of the plots by providing commands under the specific `subplot()` function.

#### Visualizing Matrix

If we have a matrix of different values and we want to visualize it then, by passing the matrix into `imagesc()` produces a color-grid for the matrix with divergent colors.

````octave
% Create a (5x5) matrix
A = magic(5);

% Producing color-grid for the matrix
imagesc(A)

% For producing the color scale applied
colorbar;
````

if we want a gradient of single color in the matrix then, we can pass the following command :

````octave
% Create a (15x15) matrix
B = magic(15);

% Producing color-grid for the matrix
imagesc(B)

% For producing grey gradient
colormap gray;

% For producing the color scale applied
colorbar;
````
